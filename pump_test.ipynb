# Runze Pump Testing Notebook

This notebook explores and tests the functions available in the RunzePump class from matterlab_pumps.

## 1. Import Required Libraries

Import the RunzePump class from matterlab_pumps and any other necessary libraries.

```python
from matterlab_pumps import RunzePump
import serial
import serial.tools.list_ports
import time
```

## 2. Detect Available COM Ports

First, let's see what serial ports are available.

```python
import serial
import serial.tools.list_ports

# List available COM ports
ports = serial.tools.list_ports.comports()
available_ports = [port.device for port in ports]
print("Available COM ports:")
for port in ports:
    print(f"  {port.device}: {port.description}")
    
if available_ports:
    com_port = available_ports[0]
    print(f"\nUsing port: {com_port}")
else:
    print("\nNo COM ports found!")
    com_port = None

com_port = "/dev/ttyUSB0"  # Update this to the correct COM port 
```

## 3. Initialize the Pump

Create a RunzePump instance with the appropriate parameters.

```python
# Initialize the pump if a COM port is available
if com_port:
    pump = RunzePump(
        com_port=com_port, 
        address=0, 
        syringe_volume=1e-3,  # 1 mL in liters
        num_valve_port=12
    )
    print(f"Pump initialized on port {com_port}")
else:
    pump = None
    print("Pump initialization skipped - no COM port available")
```

## 4. Test Basic Movement Functions

Test basic pump operations like draw with different volumes and speeds.

```python
# Test basic draw operation
if pump:
    try:
        print("Testing draw operation...")
        # Draw 0.5 mL at speed 0.5 from valve port 1
        pump.draw(volume=0.5, valve_port=1, speed=0.5)
        print("Draw operation completed successfully!")
        time.sleep(1)
    except Exception as e:
        print(f"Error during draw operation: {e}")
else:
    print("Pump not initialized, skipping draw test")
```

```python
# Inspect draw() method signature
if pump:
    import inspect
    sig = inspect.signature(pump.draw)
    print(f"draw() method signature: {sig}")
    print(f"\nDraw method docstring:")
    print(pump.draw.__doc__)
else:
    print("Pump not initialized")
```

```python
# Check key pump methods
if pump:
    import inspect
    print("Key pump methods:")
    methods = [method for method in dir(pump) if not method.startswith('_') and callable(getattr(pump, method))]
    
    # Filter for likely key methods
    key_methods = [m for m in methods if any(word in m.lower() for word in ['draw', 'dispense', 'pump', 'move', 'valve', 'initialize', 'stop', 'home'])]
    
    for method in sorted(key_methods):
        attr = getattr(pump, method)
        try:
            sig = inspect.signature(attr)
            doc = attr.__doc__.split('\n')[0] if attr.__doc__ else ""
            print(f"  {method}{sig}")
            if doc.strip():
                print(f"    → {doc.strip()}")
        except:
            print(f"  {method}()")
else:
    print("Pump not initialized")
```

```python
# Test dispense operation (push liquid OUT)
if pump:
    try:
        print("Testing dispense operation...")
        # First draw liquid from port 1
        print("  Drawing 0.5 mL from valve port 1...")
        pump.draw(volume=0.5, valve_port=1, speed=0.3)
        time.sleep(0.5)
        
        # Then dispense to port 2
        print("  Dispensing 0.5 mL to valve port 2...")
        pump.dispense(volume=0.5, valve_port=2, speed=0.3)
        print("Dispense operation completed successfully!")
    except Exception as e:
        print(f"Error during dispense operation: {e}")
else:
    print("Pump not initialized")
```

```python
# Empty the syringe first by dispensing everything
if pump:
    try:
        print("Emptying the syringe...")
        pump.dispense_all()
        print("Syringe emptied successfully!")
        time.sleep(0.5)
    except Exception as e:
        print(f"Error emptying syringe: {e}")
else:
    print("Pump not initialized")
```

```python
# Test draw_and_dispense (one step to move liquid from port A to B)
if pump:
    try:
        print("Testing draw_and_dispense operation...")
        # Draw from port 1 and dispense to port 3 in one operation
        pump.draw_and_dispense(
            volume=0.3, 
            draw_valve_port=1, 
            dispense_valve_port=3, 
            speed=0.3
        )
        print("Draw and dispense completed! Liquid moved from port 1 to port 3.")
    except Exception as e:
        print(f"Error during draw_and_dispense operation: {e}")
else:
    print("Pump not initialized")
```

## 5. Test Volume Operations

Experiment with different volume settings and verify pump behavior.

```python
# Test different volumes
if pump:
    volumes_to_test = [0.1, 0.5, 1.0]
    
    for volume in volumes_to_test:
        try:
            print(f"Testing volume: {volume} mL")
            pump.draw(volume=volume, valve_port=1, speed=0.3)
            print(f"  Successfully drew {volume} mL")
            time.sleep(0.5)
        except Exception as e:
            print(f"  Error: {e}")
else:
    print("Pump not initialized")
```

## 6. Test Valve Control

Test switching between different valve ports.

```python
# Test different valve ports
if pump:
    ports_to_test = [1, 3, 6, 12]
    
    for port in ports_to_test:
        try:
            print(f"Testing valve port: {port}")
            pump.draw(volume=0.2, valve_port=port, speed=0.3)
            print(f"  Successfully used port {port}")
            time.sleep(0.3)
        except Exception as e:
            print(f"  Error: {e}")
else:
    print("Pump not initialized")
```

## 7. Test Speed Settings

Test different speed parameters and observe behavior.

```python
# Test different speeds
if pump:
    speeds_to_test = [0.1, 0.5, 1.0]
    
    for speed in speeds_to_test:
        try:
            print(f"Testing speed: {speed}")
            pump.draw(volume=0.1, valve_port=1, speed=speed)
            print(f"  Successfully tested speed {speed}")
            time.sleep(0.3)
        except Exception as e:
            print(f"  Error: {e}")
else:
    print("Pump not initialized")
```

## 8. Check Pump Status

Query and display the current status of the pump.

```python
# Check pump status and attributes
if pump:
    print("Pump Information:")
    print(f"  Syringe Volume: {pump.syringe_volume_mL} mL")
    print(f"  Number of Valve Ports: {pump.num_valve_port}")
    print(f"  COM Port: {pump.com_port}")
    print(f"  Address: {pump.address}")
    
    # Try to access any available pump status methods
    print("\nAvailable pump methods:")
    methods = [method for method in dir(pump) if not method.startswith('_')]
    for method in methods:
        print(f"  - {method}")
else:
    print("Pump not initialized")
```
